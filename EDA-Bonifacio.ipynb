````python
def draw_polylines(points, speeds, map):
    colors = [speed_color(x) for x in speeds]
    n = len(colors)
    # Need to have a corresponding color for each point
    #print(len(points))
    #print(n)
    if n != len(points):
        raise ValueError
    i = 0
    j = 1
    curr = colors[0]
    while i < n and j < n:
        if colors[i] != colors[j]:
            line = folium.PolyLine(points[i:j], color=curr, weight=8, opacity=1)
            line.add_to(map)
            curr = colors[j]
            i = j
        j += 1
    if i < j:
        folium.PolyLine(points[i:j], color=curr, weight=8, opacity=1).add_to(map)

color=['red','orangered','tomato','darkorange','orange','gold','yellow','greenyellow','lowngreen','limegreen','green','seagreen','teal','deepskyblue','royalblue','blue']
def speed_color(speed):
    maxspeed=course['vitesse fond'].max()
    index=int(speed*15/maxspeed)
    return color[index]
````
